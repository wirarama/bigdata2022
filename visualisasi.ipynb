```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
%matplotlib inline
```

```python
penguins = sns.load_dataset("penguins")
```

```python
sns.displot(penguins, x="flipper_length_mm")
```

```python
sns.displot(penguins, x="flipper_length_mm", binwidth=1) #dibagi per perubahan, misal setial 1 mm
```

```python
sns.displot(penguins, x="flipper_length_mm", bins=20) #dibagi 20
```

```python
sns.displot(penguins, x="flipper_length_mm", bins=[170,190,195,200,220])
```

```python
sns.displot(penguins, x="flipper_length_mm", discrete=True) #setiap jumlah yg unik
```

```python
sns.displot(penguins, x="flipper_length_mm", shrink=0.8) #jarak antar bar
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species")
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="sex", multiple="dodge")
```

```python
sns.displot(penguins, x="flipper_length_mm", col="sex",hue="species",stat="probability")
```

```python
sns.displot(penguins, x="flipper_length_mm", kind="kde")
```

```python
sns.displot(penguins, x="flipper_length_mm", kind="kde", bw_adjust=.25)
```

```python
sns.displot(penguins, x="flipper_length_mm", col="sex",hue="species", kind="kde", multiple="stack")
```

```python
sns.displot(penguins, x="flipper_length_mm", kde=True,discrete=True)
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="ecdf")
```

```python
sns.displot(penguins, x="flipper_length_mm", y="bill_depth_mm",hue="species",kind="kde")
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="kde")
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", binwidth=(2, .5), cbar=True)
```
