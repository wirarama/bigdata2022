```python
from random import randint
```

```python
data = []
for i in range(100000):
    a = randint(1,50)
    b = a*((randint(1,9)/10)*1000000)
    data.append([a,b])
```

```python
print(data)
```

```python
def perbandingan(n):
    return n[1]/n[0]
```

```python
hasil = map(perbandingan,data)
```

```python
print(list(hasil))
```

```python
print(max(list(hasil)))
```
