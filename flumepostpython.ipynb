```python
import requests
```

```python
r = requests.post('http://localhost:81/', 
                  json=[{"headers" : { "host" : "localhost" },"body": "test"}])
print(r.status_code)
```

```python
for i in range (1000):
    r = requests.post('http://localhost:81/', 
                  json=[{"headers" : { "host" : "localhost" },
                         "body": "test "+str(i)}])
    print(r.status_code)
```
